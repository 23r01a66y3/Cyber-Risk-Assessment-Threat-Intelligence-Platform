```python
import json
import requests
```

```python
# Load dataset
url = "https://raw.githubusercontent.com/mitre/cti/master/enterprise-attack/enterprise-attack.json"
data = requests.get(url).json()
```

```python
# Explore structure
print("Top-level keys:", data.keys())
print("Total objects:", len(data["objects"]))
```

```python
# Extract only attack-pattern techniques
techniques = [
    obj for obj in data["objects"]
    if obj.get("type") == "attack-pattern"
]

print("Total attack techniques:", len(techniques))
```

```python
#Build Threat Scoring Logic
def threat_score(tech):
    score = 5

    name = tech.get("name", "").lower()
    description = tech.get("description", "").lower()

    # Name-based scoring
    if "credential" in name:
        score += 3
    if "execution" in name:
        score += 2
    if "privilege" in name:
        score += 3
    if "persistence" in name:
        score += 2
    if "lateral" in name:
        score += 2

    # Description-based scoring
    if "administrator" in description:
        score += 2
    if "remote" in description:
        score += 2
    if "bypass" in description:
        score += 2
    if "stealth" in description:
        score += 1

    return score
```

```python
#Score All Techniques
scored_techniques = []

for tech in techniques:
    name = tech.get("name", "Unknown Technique")
    score = threat_score(tech)
    scored_techniques.append((name, score))

print("Scoring complete. Total scored:", len(scored_techniques))
```

```python
#Prioritize Threats
scored_techniques.sort(key=lambda x: x[1], reverse=True)

print("\nTop 10 Highest-Risk Techniques:\n")

for name, score in scored_techniques[:10]:
    print(f"{name} — Score: {score}")
```

```python
#Critical Threat Filter
print("\nCritical Threats (Score >= 8.9):\n")

critical = [t for t in scored_techniques if t[1] >= 8.9]

for name, score in critical:
    print(f"{name} — Score: {score}")

print("\nTotal Critical Techniques:", len(critical))
```
